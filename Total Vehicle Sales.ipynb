```python
import pandas as pd
import matplotlib.pyplot as plt


# Output CSV File
#output_data_file = "output_data.csv"
```

```python
total_sales_df = pd.read_csv(r"C:\Users\idriv\Desktop\Total_Vehicle_Sales_(Units).csv")
total_sales_df.set_index("observation_date")
total_sales_df = total_sales_df.set_index("observation_date")
```

```python
total_sales_df.plot(kind="bar", figsize=(20,10), color="#069AF3")
plt.ylabel("Total Vehicles Sold (Millions)", fontweight="bold", fontsize=15)
plt.xlabel("", fontweight="bold", fontsize=15)
plt.title("Total Vehicles Sold 2020-2022", fontweight="bold", fontsize=20)
plt.xticks(rotation=45)

plt.savefig("total_vehicles_sold.png")

plt.show()
```

```python
sales_in_dollars_df = pd.read_csv(r"C:\Users\idriv\OneDrive\Desktop\total_vehicle_sales_in_dollars.csv")
sales_in_dollars_df.set_index("observation_date")
sales_in_dollars_df = sales_in_dollars_df.set_index("observation_date")

sales_in_dollars_df.plot(kind="bar", figsize=(20,10), color="mediumseagreen")

plt.ylabel("Dollars (Millions)", fontweight="bold", fontsize = 15)
plt.xlabel("")
plt.xticks(rotation=45)
plt.title("Total Sales Revenue 2020 - 2022", fontweight="bold", fontsize=20)

plt.savefig("total_sales_rev.png")

plt.show()
```

```python
consumer_price_index_df = pd.read_csv(r"C:\Users\idriv\OneDrive\Desktop\consumer_price_index.csv")

consumer_price_index_df = consumer_price_index_df.set_index("observation_date")
consumer_price_index_df.plot(kind="bar", figsize=(20,10), color="coral")

plt.ylabel("Index Value)", fontweight="bold", fontsize = 15)
plt.xlabel("")
plt.xticks(rotation=45)
plt.title("Consumer Price Index 2020 - 2022", fontweight="bold", fontsize=20)

plt.savefig("consumer_price_index.png")

plt.show()
```
