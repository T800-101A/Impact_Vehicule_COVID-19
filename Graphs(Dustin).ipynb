```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
producer_price = pd.read_csv("Data/Producer Price Index by Commodity Transportation Equipment Motor Vehicles.csv")
manufacture_orders = pd.read_csv("Data/Manufacturers' New Orders Motor Vehicles and Parts.csv")
invintory = pd.read_csv("Data/Domestic Auto Inventories.csv")
production = pd.read_csv("Data/Domestic Auto Production.csv")
```

```python
production = production.loc[60:]
invintory = invintory.loc[60:]
manufacture_orders = manufacture_orders.loc[59:]
producer_price = producer_price.loc[60:]
```

```python
def graph2(x,y,y_label,title):
    plt.figure(figsize=(20,10))
    plt.plot(x, y)
    plt.axvline(x = "2020-03-01", color = 'r', label = 'Stimulous Checks')
    plt.axvline(x = "2020-12-01", color = 'r')
    plt.axvline(x = "2021-03-01", color = 'r')
    plt.axvline(x = "2020-09-01", color = 'g', label = 'Chip Shortage')
    plt.xticks(rotation = 55)
    plt.ylabel(y_label, fontsize=20)
    plt.title(title,fontsize=30)
    plt.legend()
    plt.show()
```

```python
def graph1(x,y,y_label,title):
    plt.figure(figsize=(20,10))
    plt.plot(x, y)
    plt.xticks(rotation = 55)
    plt.ylabel(y_label, fontsize=20)
    plt.title(title,fontsize=30)
    plt.show()
```

```python
graph1(production["DATE"], production["DAUPSA"],"Average Output","Domestic Auto Production")
```

```python
graph1(producer_price["DATE"], producer_price["WPS1411"],"Index","Producer Price")
```

```python
graph1(manufacture_orders["DATE"], manufacture_orders["AMVPNO"],"Orders","New Orders(Motor Vehicle Parts)")
```

```python
graph2(invintory["DATE"], invintory["AUINSA"],"Invintory","Average Auto Invintory w/ Stimulus Checks")
```

```python
correlation_invintory_vs_price = st.pearsonr(producer_price["WPS1411"],invintory["AUINSA"])[0]
print(f"The correlation between the price and the invintory is {correlation_invintory_vs_price}")
```
